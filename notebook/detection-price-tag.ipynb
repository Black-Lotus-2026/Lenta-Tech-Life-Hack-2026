```python
import cv2
from ultralytics import YOLO

# --- НАСТРОЙКИ (измените эти пути) ---
MODEL_PATH = r"D:\Lenta-Tech-Life-Hack-2026\weight\best-price-tag.pt"  # Например: "best_price_tags.pt"
VIDEO_PATH = r"D:\Lenta-Tech-Life-Hack-2026\data\Данные\25_12-20\25_12-20.mp4" # Например: "video.mp4"
OUTPUT_PATH = "результат_обработки.mp4" # Например: "output.mp4"
CONFIDENCE_THRESHOLD = 0.5 # Минимальная уверенность для срабатывания (0-1)
# -------------------------------------

def main():
    # Загружаем вашу обученную модель
    model = YOLO(MODEL_PATH)

    # Открываем видеофайл
    cap = cv2.VideoCapture(VIDEO_PATH)
    if not cap.isOpened():
        print(f"Ошибка: Не удалось открыть видео {VIDEO_PATH}")
        return

    # Получаем параметры видео (ширина, высота, FPS) для сохранения результата
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Настраиваем кодек и объект для записи выходного видео
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(OUTPUT_PATH, fourcc, fps, (width, height))

    frame_count = 0
    print(f"Начинаю обработку видео '{VIDEO_PATH}'...")

    # Цикл по кадрам
    while cap.isOpened():
        success, frame = cap.read()
        if not success:
            break  # Видео закончилось

        frame_count += 1
        # Каждые 30 кадров выводим прогресс, чтобы не засорять консоль
        if frame_count % 30 == 0:
            print(f"Обработано кадров: {frame_count}")

        # Запускаем детекцию на текущем кадре
        # results = model(frame, conf=CONFIDENCE_THRESHOLD) # Более старый способ
        results = model.predict(frame, conf=CONFIDENCE_THRESHOLD, verbose=False)

        # Визуализируем результаты прямо на кадре
        # Благодаря параметру boxes=False мы не рисуем рамки отдельно, а используем встроенную аннотацию
        annotated_frame = results[0].plot()

        # Записываем аннотированный кадр в выходной файл
        out.write(annotated_frame)

        # ОПЦИОНАЛЬНО: Показать видео в реальном времени.
        # Нажмите 'q', чтобы прервать обработку досрочно.
        # cv2.imshow("Detect Price Tags", annotated_frame)
        # if cv2.waitKey(1) & 0xFF == ord("q"):
        #     break

    # Освобождаем ресурсы
    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print(f"✅ Готово! Результат сохранен в {OUTPUT_PATH}")

if __name__ == "__main__":
    main()
```
