```python
import cv2
import numpy as np
from pyzbar.pyzbar import decode

# 1. Загружаем изображение с QR-кодом
# Укажите путь к вашему файлу
image_path = "../outputs/extracted_price_tags_25_12_20/img_1.png"
img = cv2.imread(image_path)

# 2. Пытаемся найти и расшифровать QR-код(ы) на изображении
decoded_objects = decode(img)

# 3. Обрабатываем результат
if decoded_objects:
    for obj in decoded_objects:
        # Текст, закодированный в QR-коде
        print(f"Расшифрованные данные: {obj.data.decode('utf-8')}")
else:
    print("QR-код не найден.")

# Если нужно показать изображение с найденным QR-кодом (раскомментируйте)
for obj in decoded_objects:
    # Рисуем прямоугольник вокруг QR-кода
    pts = np.array(obj.polygon)
    if len(pts) == 4:
        cv2.polylines(img, [pts.astype(int)], True, (0, 255, 0), 2)
#
cv2.imshow("Result", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
